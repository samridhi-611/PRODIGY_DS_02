Cleaning Data

```python

missing_data = data.isnull().sum()
missing_data
```

```python
data = data.drop_duplicates()
data.head(8)
```

```python

data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Fare'] = pd.to_numeric(data['Fare'], errors='coerce')
```

Exploratory Data Analysis (EDA)

```python
summary_stats = data.describe()
print(summary_stats)
```

Univariate Analysis:

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data['Age'].dropna(), kde=True , color="olive")
plt.title('Age Distribution')
plt.subplot(1, 2, 2)
sns.countplot(data['Sex'],color="steelblue")
plt.title('Gender Distribution')
plt.show()
```

Categorical Variables:

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(data['Pclass'],color="deeppink")
plt.title('Pclass Distribution')
plt.subplot(1, 2, 2)
sns.countplot(data['Embarked'],color="indigo")
plt.title('Embarked Distribution')
plt.show()
```

Survival Analysis:

```python
survival_rate_by_class = data.groupby('Pclass')['Survived'].mean()
survival_rate_by_sex = data.groupby('Sex')['Survived'].mean()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=survival_rate_by_class.index, y=survival_rate_by_class.values , color="aquamarine")
plt.title('Survival Rate by Pclass')
plt.subplot(1, 2, 2)
sns.barplot(x=survival_rate_by_sex.index, y=survival_rate_by_sex.values, color="crimson")
plt.title('Survival Rate by Sex')
plt.show()
```

Visualizing Survival by Age:

```python
plt.figure(figsize=(9, 6))
sns.boxplot(x='Survived', y='Age', data=data)
plt.title('Survival by Age')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
data['FamilySize'] = data['SibSp'] + data['Parch']
```

```python
sns.pairplot(data, vars=['Age', 'Fare', 'FamilySize'], hue='Survived')
plt.suptitle('Pair Plot of Age, Fare, and FamilySize')
plt.show()
```

```python
plt.figure(figsize=(9, 6))
data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.')
sns.countplot(data['Title'],color="firebrick")
plt.title('Passenger Titles')
plt.xticks(rotation=45)
plt.show()
```

```python
# Survival by Title:
title_survival_rate = data.groupby('Title')['Survived'].mean()
plt.figure(figsize=(12, 5))
sns.barplot(x=title_survival_rate.index, y=title_survival_rate.values,color="teal")
plt.title('Survival Rate by Title')
plt.xticks(rotation=45)
plt.show()
```
